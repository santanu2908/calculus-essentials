# Plotting f(x) = x² and its Derivative

This notebook plots a function and its derivative side by side using **matplotlib**.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Define the function and its derivative

We use `np.linspace` to create 300 evenly spaced points from -3 to 3.

```python
x = np.linspace(-3, 3, 300)  # Generate 300 points between -3 and 3
f = x ** 2        # f(x) = x²
f_prime = 2 * x   # f'(x) = 2x
```

## Plot side by side

`plt.subplots(1, 2)` creates one row of two axes so we can compare them.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4)) # Create a figure with 1 row and 2 columns of subplots

# Left plot — the original function
ax1.plot(x, f, color='#2563eb', linewidth=2)
ax1.set_title('f(x) = x²', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.axhline(0, color='grey', linewidth=0.5)
ax1.axvline(0, color='grey', linewidth=0.5)
ax1.grid(True, alpha=0.3)




# Right plot — the derivative
ax2.plot(x, f_prime, color='#dc2626', linewidth=2)
ax2.set_title("f'(x) = 2x", fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.axhline(0, color='grey', linewidth=0.5)
ax2.axvline(0, color='grey', linewidth=0.5)
ax2.grid(True, alpha=0.3)


fig.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

## Key takeaways

| Observation | Why? |
|---|---|
| The derivative crosses **zero at x = 0** | That's where x² has its minimum (slope = 0) |
| 2x is **negative** for x < 0 | x² is decreasing on the left side |
| 2x is **positive** for x > 0 | x² is increasing on the right side |
